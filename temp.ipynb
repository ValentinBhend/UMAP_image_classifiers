```python
from scipy.optimize import minimize
import numpy as np
```

```python
nn = [.5, .5, 1, 2]
l2 = np.log2(5)

sigm = 1

def snn(sigm, nn, l2):
    rho = nn[0]
    ret = 0
    for n in nn:
        ret += np.exp(-(n - rho) / sigm)
    return (ret - l2)**2

res = minimize(snn, x0=[1], args=(nn,l2))
sigma = res.x
print(sigma)
rho = nn[0]
for n in nn:
    print(np.exp(-(n - rho) / sigma))

print(snn(sigma, nn, l2))
```
